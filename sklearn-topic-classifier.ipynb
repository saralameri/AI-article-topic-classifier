```python
import numpy as np
import pandas as pd
import os

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split 
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import recall_score, precision_score, f1_score, confusion_matrix, roc_curve, auc

from nltk.tokenize import wordpunct_tokenize
#, sent_tokenize, word_tokenize 
from nltk.stem import PorterStemmer
```

```python

#creating the traning set data frame 
articles_set = pd.DataFrame(columns = ["article", "topic"])
```

```python

#creating a new Porter stemmer object
Porter = PorterStemmer()
def stemSentence(sentence):
    #divide the sting into a list of words
    #useing the 
    token_words = wordpunct_tokenize(sentence)
    #create a list to store the words afterstemmiing them
    stem_sentence=[]
    #go throgh all the words in the sentence 
    for word in token_words:
        #addint the stemed word to the stem_sentence
        stem_sentence.append(Porter.stem(word))
        #adding a space after it
        #(so the words will have spaces between them in the sentence)
        stem_sentence.append(" ")
    #turn the list into a sentence and return it 
    return "".join(stem_sentence)
```

```python
#reading the traning articles and adding them to the traning set

#reading the folders(topic) in the training folder 
articles = os.listdir("articles") 

#for each topic folder in the traning folder
for topic in articles :
    #reading all the files(articles) in this topic
    topic_articles = os.listdir("articles"+"/"+topic)

    #go throgh each article in this topic folder
    for article in topic_articles:
        #opening this article txt file
        file = open("articles/"+topic+"/"+article,'r', encoding='utf-8')
        #reading to get the artical as a string
        file_string = file.read()
        #to stem the 
        file_string = stemSentence(file_string)

        #to add a new row to the datafram using concat
        #first we creat the row
        new_row = [file_string, topic]
        #then we create a dataframe with only that row
        new_df = pd.DataFrame([new_row], columns = ["article", "topic"])
        #then concat it with the traning set data frame 
        articles_set = pd.concat([articles_set, new_df], ignore_index = True)

        #closing the file
        file.close()


#training_set.to_string()
articles_set
```

```python
vectorizer = CountVectorizer(stop_words='english')
#creating a document-term matrix and getting the vocab
all_features = vectorizer.fit_transform(articles_set.article)
#number of columns = total number of words in the whole traning set
#number of rows = number of artical in the traning set
print(all_features.shape)
all_features.toarray()
```

```python
vectorizer.vocabulary_
```

```python
article_train, article_test, lable_train, lable_test = train_test_split(all_features, articles_set.topic, test_size=0.1, random_state=88)
article_train.shape, article_test.shape
```

```python
classifier = MultinomialNB()
#traning 
classifier.fit(article_train, lable_train)
```

```python
art_test
```

```python
classifier.predict_proba(article_test)
```

```python
nr_correct = (lable_test == classifier.predict(article_test)).sum()
print(f"articles classified correctly: {nr_correct}")

nr_incorrect = lable_test.size - nr_correct
print(f"articles classified incorrectly: {nr_incorrect}")
```

```python
accuracy = classifier.score(article_test, lable_test)
print(f"the testing accuracy of the model is {accuracy:.2%}")
```

```python
#dividing the true positives by anything that was predicted as a positive. 
#out of all sport articles classified how many were classified corectlly 
precision_score(lable_test, classifier.predict(article_test), pos_label = 'sports', average = 'binary')
```

```python
#recall(True Positive Rate) 
#dividing the true positives by anything that should have been predicted as positive
#out of all sport articles how many were classified as sport
recall_score(lable_test, classifier.predict(article_test), pos_label = 'sports', average = 'binary')
```

```python
#the balance between the precision and the recall
#2 * ( (precision*recall) / (precision + recall) )
#the 2 is wight
f1_score(lable_test, classifier.predict(article_test), pos_label = 'sports', average = 'binary')
```

```python
#creating the confusion matrix
confusion_matrix(lable_test, classifier.predict(article_test), labels = ['sports', 'technology'], )
```
